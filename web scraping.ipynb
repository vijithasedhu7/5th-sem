```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
import requests
```

```python
url="https://www.jiomart.com/c/fashion/women/watches/4121"
```

```python
res=requests.get(url)
```

```python
res
```

```python
#pip install --upgrade pandas --user
```

```python
Back_code=BeautifulSoup(res.text)
```

```python
Back_code
```

```python
product_name=Back_code.select(" .jm-fc-primary-grey-80")
```

```python
product_name
```

```python
price=Back_code.select(".jm-heading-xxs ")
```

```python
price
```

```python
offer=Back_code.select(".jm-badge ")
```

```python
offer
```

```python

Mobile_data=pd.DataFrame(product_name,columns=["Name"])
```

```python
Mobile_data["price"]=price
```

```python
Mobile_data["offer"]=offer
```

```python
Mobile_data
```

```python
Mobile_data.to_csv("WATCHES.csv")
```
